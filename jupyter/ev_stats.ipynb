```python
import pandas as pd
```

```python
stat_file = "ev_stats.csv"
data = pd.read_csv(stat_file, decimal=",", header=0, names=["Date", "Consumption", "Distance", "Time hh:mm", "Speed"], index_col=0, parse_dates=True)
data['Duration'] = data['Time hh:mm'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]))
data = data.drop(columns=['Time hh:mm']).loc[(data['Speed']>30) & (data['Duration'] > 15)]
data.head()
```

```python
print(data.shape)
print(data.dtypes)
data.describe()
```

```python
data['Consumption'].plot(figsize=(12,8))
```
